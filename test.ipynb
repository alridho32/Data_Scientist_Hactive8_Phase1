```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['High', 'Medium', 'Jam', 'Low']
sizes = [10, 24.8, 32.1, 33]
colors = ['#ff9999', '#F05A28', '#ffd699', '#97ffca']  # Pastel shades for High, Medium, Jam, Low

# Create the pie chart with pastel colors and add a border for separation
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90,
                                  wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},  # Adding black border to each segment
                                  textprops={'fontsize': 12, 'color': 'black'})
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Set the background color to orange
fig.patch.set_facecolor('#F6921E')  # Light orange background

# Save the chart as an image
plt.savefig('pie_chart_pastel_orange_bg_with_border_updated.png', bbox_inches='tight', facecolor=fig.get_facecolor())

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
agent_ages = [20, 22, 25, 27, 30, 32, 35, 37] * 10
agent_ratings = [5.0] * 30 + [4.0] * 30 + [3.5] * 20 + [3.0] * 10 + [2.5] * 10

# Create the scatter plot
fig, ax = plt.subplots()

# Plotting Very Good/Good ratings (>= 4.0) in blue
very_good_x = [age for age, rating in zip(agent_ages, agent_ratings) if rating >= 4.0]
very_good_y = [rating for rating in agent_ratings if rating >= 4.0]
ax.scatter(very_good_x, very_good_y, color='blue', label='Very Good/Good', alpha=0.7)

# Plotting Others ratings (< 4.0) in red
others_x = [age for age, rating in zip(agent_ages, agent_ratings) if rating < 4.0]
others_y = [rating for rating in agent_ratings if rating < 4.0]
ax.scatter(others_x, others_y, color='red', marker='x', label='Others', alpha=0.7)

# Set chart title and labels
ax.set_title('Agent Age vs Agent Rating and Tingkat Rating')
ax.set_xlabel('Agent Age')
ax.set_ylabel('Agent Rating')

# Add legend
ax.legend(title='Tingkat Rating')

# Set the background color to #F6921E
fig.patch.set_facecolor('#F6921E')

# Save the chart as an image
plt.savefig('scatter_plot_custom_bg.png', bbox_inches='tight', facecolor=fig.get_facecolor())

# Show the chart
plt.show()
```
